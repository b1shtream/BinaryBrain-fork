```python
import matplotlib.pyplot as plt
import binarybrain as bb
import cv2
```

```python
def plot_cifar10(x):
    img = cv2.merge(x.reshape(3, 32, 32))
    plt.imshow(img)

def plot_cifar10_list(x0, x1, n, o=0):
    plt.figure(figsize=(10, 4))
    for i in range(n):
        plt.subplot(2, n, i+1)
        plot_cifar10(x0[i+o])
        plt.subplot(2, n, n+i+1)
        plot_cifar10(x1[i+o])
```

```python
td = bb.load_cifar10()
x = td['x_train'].copy()
x_shape = td['x_shape']
```

```python
ida = bb.image_data_augmentation(shift_x_range=0.2, shift_y_range=0.2, flip_x_rate=0.5, rotation_range=5.0, scale_range=0.2, rate=0.5)
```

```python
x, t = ida(td['x_train'].copy(), td['t_train'].copy(), td['x_shape'], td['t_shape'])
plot_cifar10_list(td['x_train'], x, 8)
```
